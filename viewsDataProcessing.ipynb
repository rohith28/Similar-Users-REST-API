In this Notebook I am trying to process the data and save it in MySQL ddatabase. As each user has different session for same course_id we need to add them. 

```python
import pandas as pd
```

```python
courseViewData = pd.read_csv('data_files_ml_engineer/user_course_views.csv')
```

```python
courseViewData.head()
```

```python
usersHandle = courseViewData['user_handle'].unique()
usersHandle
```

```python
resultDF = pd.DataFrame(columns=['user_handle','course_id','level','view_time_seconds'])
```

```python
for index in usersHandle:
    courses = courseViewData['course_id'].loc[courseViewData['user_handle'] == index]
    print(index)
    for course in courses:
        tempDF = pd.DataFrame(columns=['user_handle','course_id','level','view_time_seconds'])
        df = courseViewData.loc[(courseViewData['user_handle']== index) &(courseViewData['course_id']== course)]
        
        if course in resultDF['course_id'].values :
            pass
        else:
            tempDF['user_handle'] = df['user_handle'].unique()
            tempDF['course_id'] = df['course_id'].unique()
            tempDF['level'] = df['level'].unique()
            tempDF['view_time_seconds'] = df['view_time_seconds'].sum()
            resultDF = resultDF.append(tempDF)
            print(tempDF)
```

## Label Encoder

Using label encoder to change the categorical value to numerical value

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(["Beginner","Intermediate","Advanced"])
```

```python
levelDF = pd.DataFrame(le.transform(resultDF['level']))
```

```python
resultDF['level'] = levelDF.values
```
